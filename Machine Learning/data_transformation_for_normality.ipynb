## Try This If Your Linear Regression Model is Underperforming

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

import seaborn as sns
sns.set()
```

```python
color_code = '#fe7c73'
```

```python
data = np.random.gamma(2, 1, 1000) * 10
```

```python
plt.hist(data, bins=30, edgecolor='black', color = color_code)
plt.title('Original Data Distribution', fontsize = 20, fontweight = "bold");
```

```python
# Apply logarithmic transformation
data_log = np.log(data)

# Plot log-transformed data distribution
plt.hist(data_log, bins=30, edgecolor='black', color = color_code)
plt.title('Log-Transformed Data', fontsize = 20, fontweight = "bold");
```

```python
# Apply square root transformation
data_sqrt = np.sqrt(data)

# Plot square root-transformed data distribution
plt.hist(data_sqrt, bins=30, edgecolor='black', color = color_code)
plt.title('Square Root-Transformed Data', fontsize = 20, fontweight = "bold");
```

```python
# Apply Box-Cox transformation
data_boxcox, _ = stats.boxcox(data)

# Plot Box-Cox transformed data distribution
plt.hist(data_boxcox, bins=30, edgecolor='black', color = color_code)
plt.title('Box-Cox Transformed Data', fontsize = 20, fontweight = "bold");
```

## My view

At times, skewness in data can negatively influence the predictive power of your linear regression model.

If the model is underperforming, it may be due to skewness. There are many ways to handle this:
- Try a different model
- Transform the skewed variable

The most commonly used transformations are:
- Log transform: Apply the logarithmic function.
- Sqrt transform: Apply the square root function.
- Box-cox transform: You can use it using Scipy’s implementation: [Scipy docs](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.boxcox.html).

The effectiveness of the three transformations is clear from the image above. Applying them transforms the skewed data into a (somewhat) normally distributed variable.
