## Multiplication

```python
def mul(x,y):
    return x*y
mul(2,8)
```

```python
mul = lambda x, y:x*y
mul(2,8)
```

```python
(lambda x, y:x*y)(2,8)
```

## Sort a list of tuples based on their alphabets

```python
tuples = [(1,'d'), (2,'c'), (3,'b'), (4,'a')]
sorted(tuples, key=lambda x:x[1])
```

```python
import matplotlib.pyplot as plt
nums = plt.plot(sorted(range(-100,101), key=lambda x: x*x))
plt.show()
```

## Nested Function and lambda

```python
def power(n):
    return lambda x: x**n
power_3 = power(3)
list(map(power_3, [2,3,4,5]))
```

## Manipulate DataFrame

```python
import pandas as pd
df = pd.DataFrame([[1,2,3],[4,5,6]], columns=['a','b','c'])
df
```

```python
# create the 4th columns that is the sumn of the other 3 columns 

df['d'] = df.apply(lambda row:row['a'] + row['b'] + row['c'], axis=1)
df
```
