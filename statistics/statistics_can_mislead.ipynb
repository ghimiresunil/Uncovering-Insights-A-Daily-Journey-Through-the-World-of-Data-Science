## Why Correlation (and Other Statistics) Can Be Misleading.

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df_with_outliers = pd.read_csv("../dataset/Correlation_example.csv")
df_with_outliers.head()
```

```python
df_without_outliers = df_with_outliers.loc[:47]
```

```python
df_without_outliers.plot(kind='scatter',x='x', y='y')
```

```python
sns.lmplot(x = "x", y = "y", data = df_without_outliers, ci = None)
```

```python
sns.lmplot(x = "x", y = "y", data = df_with_outliers, ci = None)
```
