## When using summary statistics, be caution before making any conclusions.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles, make_blobs
```

```python
fig, ax = plt.subplots(3, 3, figsize=(16, 12))

X = np.linspace(0, 3 * np.pi, num=100)
Y = np.sin(X) + np.random.normal(0, 0.1, size=X.shape)
ax[0][0].scatter(X, Y, s=25, color='blue')

X = np.linspace(-5, 5, num=100)
Y = X ** 2 + np.random.normal(0, 1, size=X.shape)
ax[0][1].scatter(X, Y, s=25, color='red')

X, y = make_circles(n_samples=200, noise=0.04, random_state=42, factor=0.5)
ax[0][2].scatter(X[:, 0], X[:, 1], s=25, color='green')

X_inner = X[y == 1]
ax[2][0].scatter(X_inner[:, 0], X_inner[:, 1], s=25, color='orange')

X = np.linspace(1, 10, num=100)
Y = np.log(X) + np.random.normal(0, 0.05, size=X.shape)
X = np.concatenate((-X[::-1], X))
Y = np.concatenate((Y[::-1], Y))
ax[1][1].scatter(X, Y, s=25, color='purple')

X = np.linspace(-3, 3, num=100)
Y = np.exp(-X ** 2) + np.random.normal(0, 0.03, len(X))
ax[1][2].scatter(X, Y, s=25, color='brown')

centers = [(2, 2), (-2, -2), (-2, 2), (2, -2)]
X, y = make_blobs(n_samples=200, centers=centers, cluster_std=0.3)
ax[1][0].scatter(X[:, 0], X[:, 1], s=25, color='gray')

X = np.linspace(0, 4, num=100)
Y = np.concatenate([np.zeros(20), np.ones(20), np.zeros(20), np.ones(20), np.zeros(20)]) + np.random.normal(0, 0.05,
                                                                                                              size=X.shape)
ax[2][1].scatter(X, Y, s=25, color='pink')

X = np.linspace(-1, 1, num=100)
Y = np.abs(X + np.random.normal(0, 0.03, size=X.shape))
ax[2][2].scatter(X, Y, s=25, color='cyan')

# Remove x-axis labels
for i in range(3):
    for j in range(3):
        ax[i][j].set_xticklabels([])

# Remove y-axis labels
for i in range(3):
    for j in range(3):
        ax[i][j].set_yticklabels([])

# Set titles for each subplot with 'DejaVu Sans' font
ax[0][0].set_title("Sinusoid", fontsize=17, fontweight="bold", fontfamily="DejaVu Sans")
ax[0][1].set_title("Quadratic", fontsize=17, fontweight="bold", fontfamily="DejaVu Sans")
ax[0][2].set_title("Concentric Circles", fontsize=17, fontweight="bold", fontfamily="DejaVu Sans")
ax[1][0].set_title("Blobs", fontsize=17, fontweight="bold", fontfamily="DejaVu Sans")
ax[1][1].set_title("Symmetric Log", fontsize=17, fontweight="bold", fontfamily="DejaVu Sans")
ax[1][2].set_title("Bell Curve", fontsize=17, fontweight="bold", fontfamily="DejaVu Sans")
ax[2][0].set_title("Circle", fontsize=17, fontweight="bold", fontfamily="DejaVu Sans")
ax[2][1].set_title("Step function", fontsize=17, fontweight="bold", fontfamily="DejaVu Sans")
ax[2][2].set_title("Absolute Value", fontsize=17, fontweight="bold", fontfamily="DejaVu Sans")

plt.show()
```
